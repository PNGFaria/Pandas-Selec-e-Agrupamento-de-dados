```python
pop = dados[dados['POPULAÇÃO'].str.contains('\(', na= False)]
pop

#A linha de código que você forneceu está sendo usada para manipular uma coluna chamada "POPULAÇÃO" em um DataFrame chamado "dados". Vou explicar o que essa linha faz em detalhes:

#dados['POPULAÇÃO']: Isso acessa a coluna "POPULAÇÃO" do DataFrame "dados".

#.str.contains('\(', na=False): O método .str.contains() é usado para verificar se cada valor na coluna "POPULAÇÃO" contém o caractere '('. O argumento na=False é usado para tratar valores ausentes (NaN) como "False". Isso significa que se um valor for NaN, ele será tratado como se não contivesse o caractere '('.

#[...]: O resultado da verificação anterior é usado para indexar o DataFrame "dados". Isso significa que apenas as linhas onde a coluna "POPULAÇÃO" contém o caractere '(' serão incluídas no resultado.

#Portanto, a linha de código que você forneceu está filtrando o DataFrame "dados" para incluir apenas as linhas onde a coluna "POPULAÇÃO" contém o caractere '('. Essa pode ser uma etapa útil para tratar valores que contêm parênteses antes de tentar convertê-los para números inteiros, pois os parênteses podem causar erros de conversão."""
```

```python
dados = dados.assign(populaçãoSemParenteses = dados['POPULAÇÃO'].replace('\(\d{1,2}\)', '', regex=True),
                     populaçao = lambda x : x.loc[:,'populaçãoSemParenteses'].replace('\.', '', regex = True))
```
